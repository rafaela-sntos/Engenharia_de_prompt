### Organizador de Arquivos por Extensão

Este script Python organiza arquivos em um diretório especificado, movendo-os para subpastas nomeadas com base em suas extensões. Por exemplo, todos os arquivos `.pdf` serão movidos para uma pasta `PDF`.

```
import os
import shutil

def organizar_arquivos_por_extensao(diretorio_origem):
    if not os.path.exists(diretorio_origem):
        print(f"O diretório de origem '{diretorio_origem}' não existe.")
        return

    print(f"Organizando arquivos no diretório: {diretorio_origem}")

    for filename in os.listdir(diretorio_origem):
        if os.path.isfile(os.path.join(diretorio_origem, filename)):
            file_extension = os.path.splitext(filename)[1].lower()

            # Ignora arquivos sem extensão ou arquivos ocultos (que começam com '.')
            if not file_extension or file_extension == '.':
                continue

            # Remove o ponto da extensão para o nome da pasta
            folder_name = file_extension[1:].upper()
            destino_pasta = os.path.join(diretorio_origem, folder_name)

            # Cria a pasta de destino se ela não existir
            if not os.path.exists(destino_pasta):
                os.makedirs(destino_pasta)
                print(f"Pasta '{folder_name}' criada.")

            origem_arquivo = os.path.join(diretorio_origem, filename)
            destino_arquivo = os.path.join(destino_pasta, filename)

            try:
                shutil.move(origem_arquivo, destino_arquivo)
                print(f"Movido: '{filename}' para '{folder_name}'")
            except shutil.Error as e:
                print(f"Erro ao mover '{filename}': {e}")
            except Exception as e:
                print(f"Ocorreu um erro inesperado ao mover '{filename}': {e}")

    print("Organização concluída!")
```
